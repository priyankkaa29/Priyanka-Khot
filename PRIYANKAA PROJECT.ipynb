INFORMATION : 

The Tourist.csv dataset contains information about tourist places in various cities.

The City column tells you the name of the city where the tourist place is located.

The Place column gives the name of the tourist attraction.

The Ratings_x column shows one set of ratings given to the place, likely from user reviews.

The Distance column represents how far the place is from a reference point, such as the city center.

The Place_desc column provides a description of the tourist place.

The Ratings_y column contains another rating for the same place, possibly from a different source.

The Ideal_duration column suggests how much time a tourist should ideally spend at that place.

The Best_time_to_visit column tells the best season or months to visit that location.

The City_desc column gives a short description of the city itself.

The Category column specifies the type of tourist place, such as a fort, beach, temple, museum, or natural spot.

MATPLOTLIB

1) Bar Plot 

QN) Show top 5 places with highest Ratings_x in a bar chart.

```python
top5 = data[['Place', 'Ratings_x']].sort_values('Ratings_x', ascending=False).head(5)
plt.figure(figsize=(8,5))
plt.bar(top5['Place'], top5['Ratings_x'], color='orange')
plt.title('Top 5 Highest-Rated Places')
plt.xlabel('Place')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Explanaton :The graph shows the top 5 tourist places with the highest Ratings_x in the dataset.
It visually compares how well-rated these places are, helping you quickly see which places tourists rated best and how their ratings differ.

2) Pie Chart

QN) Show how many places belong to each City using a pie chart, but display only top 10 cities.

```python
top10_cities = data['City'].value_counts().head(10)
plt.figure(figsize=(8,8))
plt.pie(top10_cities, labels=top10_cities.index, autopct='%1.1f%%', startangle=90)
plt.title('Top Cities by Number of Places')
plt.axis('equal')
plt.show()
```

Explanation : This pie chart displays the top 10 cities that have the most tourist places listed in the dataset.
Each slice of the pie shows the percentage share of places in each city, helping you visually compare which cities are most popular for tourism based on the number of attractions.

3) Line plot

QN) Compare Ratings_x and Ratings_y (average) for each city.

```python
grouped = data.groupby('City')[['Ratings_x', 'Ratings_y']].mean().head(6)
plt.plot(grouped.index, grouped['Ratings_x'], label='Ratings_x', marker='o')
plt.plot(grouped.index, grouped['Ratings_y'], label='Ratings_y', marker='x')
plt.xticks(rotation=45)
plt.title('Ratings Comparison by City')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.legend()
plt.tight_layout()
plt.show()
```

Explanation : This line plot compares the average Ratings_x and Ratings_y for the first six cities in the dataset.
It shows how these two rating types differ or match across cities, helping to understand the variation in ratings given by different sources or criteria

4) Box Plot

QN) Distribution of tourist place ratings (Ratings_x) grouped by the best time to visit (Best_time_to_visit).

```python
data.boxplot(column='Ratings_x', by='Best_time_to_visit', grid=False, rot=30)
plt.title('Ratings_x by Best Time to Visit')
plt.suptitle('')
plt.ylabel('Ratings_x')
plt.show()
```

Explanation : This box plot shows the distribution of Ratings_x for tourist places categorized by their Best Time to Visit.
Each box represents the range of ratings for places best visited in a specific time period.The thick line inside each box is the median rating.

5) Histogram

QN) Compare Ratings_x for top 2 cities using overlapping histograms.

```python
top2 = data['City'].value_counts().head(2).index
for city in top2:
    plt.hist(data[data['City'] == city]['Ratings_x'].dropna(), 
             bins=10, alpha=0.6, label=city)
plt.title('Ratings_x Comparison (Top 2 Cities)')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.legend()
plt.show()
```

Explanation : This histogram compares the Ratings_x distribution of tourist places in Shimla (blue) and Leh Ladakh (orange).
Each bar shows the number of places in each city that fall within a specific rating range (e.g., 3.2–3.4, 3.6–3.8, etc.).
Shimla has a high concentration of places rated around 3.6, indicating most places have similar ratings.
Leh Ladakh has a wider spread, with many places rated between 3.2 and 4.4, showing more variation in quality.

SEABORN

1) Bar plot

QN) How does average Ideal_duration vary by category?

```python
sns.barplot(data=data, x='Category', y='Ideal_duration', estimator='mean', ci=None)
plt.title('Average Ideal Duration by Category')
plt.xticks(rotation=45)
plt.show()
```

Explanation : T:This bar chart shows the average ideal time tourists usually spend at different types of places (categories).
X-axis: Place categories (like Nature, Temple, Beach, etc.)
Y-axis: Average ideal duration (like 2–4 hours, 3–7 hours, etc.)

2) Scatter plot

QN) Check how Ideal_duration vs Ratings_y changes by Best_time_to_visit.

```python
sns.scatterplot(data=data, x='Ideal_duration', y='Ratings_y', hue='Best_time_to_visit')
plt.title('Ideal Duration vs Ratings_y by Best Time to Visit')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Explanation : This scatter plot shows how much time people usually spend at a tourist place (Ideal Duration) and what rating they give (Ratings_y), based on the best time to visit.
Each dot is a tourist place.
X-axis: How many hours or days people stay.
Y-axis: Rating of the place.

3) Histogram

QN) What is the distribution of ratings across all products? 

```python
top2 = data['City'].value_counts().head(4).index
subset = data[data['City'].isin(top2)]
sns.kdeplot(data=subset, x='Ratings_x', hue='City', fill=True)
plt.title('Ratings_x Distribution (Top 2 Cities)')
plt.show()
```

Explanation : KDE plot to show the distribution (how ratings are spread) of the Ratings_x for each of these cities. Each city is shown in a different color with shaded areas under the curves.
The title says it shows the distribution of Ratings_x for the top cities.

4. Count Plot

QN) What is the count of tourist places for last 5 cites?

```python
sns.countplot(data=data, y='City', order=data['City'].value_counts().tail().index)
plt.title('Count of Tourist Places by City')
plt.show()
```

Explanation :This is Count Plot, In this it creates a horizontal count plot showing the number of tourist places for the cities with the fewest entries in the dataset.
